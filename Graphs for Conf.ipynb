```python
data = pd.read_csv('lbm_data/Noise_no_pulse_lbm_data_20.csv', header =None)
chosen_data = data.iloc[0,:1000]
data = chosen_data.iloc[1:1000]

y = data - 1.
print('Mean density fluctuations around : rho = {0}'.format(np.mean(y)))
fig, ax = plt.subplots(nrows=2, figsize=(20, 12))

ax[0].set_title("Signal")
ax[0].plot(np.arange(len(y)), y)
ax[0].set_xlabel(u'Iteration number', fontsize = 15)
ax[0].set_ylabel(u'Dencity fluctuation', fontsize = 15)

dt = 0.001  # sampling interval
Fs = 1 / dt  # sampling frequency

ax[1].set_title("Magnitude Spectrum")
ax[1].magnitude_spectrum(y, Fs=Fs)
ax[1].set_xlim([0, 300])

print('Thrombus status = {0}'.format(chosen_data[0]))

for i in range(2):
    ax[i].grid('on')

#plt.savefig('pulsative_flow_example.png')
plt.show()
```
